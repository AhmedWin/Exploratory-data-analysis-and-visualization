***Importing necessary libraries***

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("/kaggle/input/emp123/emp123.csv")
data.head()
```

***DataType conversion from Object to Datetime***

```python
data['Start Date'] = pd.to_datetime(data['Start Date'])
```

***Viewing the unique columns***

```python
data.nunique()
```

***Handling missing values***

```python
data.isnull().sum()
```

***Filling the missing values***

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data['Gender'].fillna("No Gender", inplace = True)
data.isnull().sum()
```

```python
mode = data['Senior Management'].mode().values[0]
data['Senior Management'] = data['Senior Management'].replace(np.nan, mode)
data.isnull().sum()
```

***Dropping columns***

```python
data = data.dropna()
data.isnull().sum()
```

***Data Encoding***

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
```

***Data Visualization***

1. Histogram

```python
sns.histplot(x = 'Salary', data = data)
```

2. Boxplot

```python
sns.boxplot(x = 'Salary', y = 'Team', data = data)
```

3. Scatterplot

```python
sns.scatterplot(x = 'Salary', y = 'Team', data = data, hue = 'Gender', size = 'Bonus %')
plt.legend(bbox_to_anchor = (1,1), loc = 2)
```

```python
sns.pairplot(data, hue = 'Gender', height = 2)
```
